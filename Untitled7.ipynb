```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    accuracy_score,
    confusion_matrix,
    classification_report,
    f1_score
)
```

```python
encoded_df = df.copy()
```

```python
label_encoder = LabelEncoder()

for column in encoded_df.select_dtypes(include="object").columns:
    encoded_df[column] = label_encoder.fit_transform(encoded_df[column])
```

```python
# Display first five rows
encoded_df.head()
```

```python
encoded_df.info()
```

```python
# Define Features and Target Variable
X = encoded_df.drop("y", axis=1)

y = encoded_df["y"]

print("Features Shape :", X.shape)

print("Target Shape :", y.shape)
```

```python
# Split Dataset into Training and Testing Sets
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("Training Features :", X_train.shape)

print("Testing Features :", X_test.shape)

print("Training Labels :", y_train.shape)

print("Testing Labels :", y_test.shape)
```

```python
# Build Logistic Regression Model
logistic_model = LogisticRegression(max_iter=1000)

logistic_model.fit(X_train, y_train)
```

```python
# Predict Test Dataset
y_pred = logistic_model.predict(X_test)
```

```python
# Calculate Model Accuracy
accuracy = accuracy_score(y_test, y_pred)

print("Accuracy :", round(accuracy * 100, 2), "%")
```

```python
# Calculate F1 Score
f1 = f1_score(y_test, y_pred)

print("F1 Score :", round(f1, 4))
```

```python
# Display Confusion Matrix
cm = confusion_matrix(y_test, y_pred)

print(cm)
```

```python
# Plot Confusion Matrix
plt.figure(figsize=(6,5))

sns.heatmap(
    cm,
    annot=True,
    fmt="d",
    cmap="Blues"
)

plt.title("Confusion Matrix - Logistic Regression")

plt.xlabel("Predicted Label")

plt.ylabel("Actual Label")

plt.show()
```

```python
print(classification_report(y_test, y_pred))
```

```python
# Display Sample Predictions
prediction_df = pd.DataFrame({
    "Actual": y_test.values,
    "Predicted": y_pred
})

prediction_df.head(10)
```

```python
# Import Required Libraries
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_curve, roc_auc_score
```

```python
# Build Random Forest Classifier
random_forest = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)

# Train the model
random_forest.fit(X_train, y_train)
```

```python
# Predict Test Dataset
rf_predictions = random_forest.predict(X_test)
```

```python
# Calculate Random Forest Accuracy
rf_accuracy = accuracy_score(y_test, rf_predictions)

print("Random Forest Accuracy :", round(rf_accuracy * 100,2),"%")
```

```python
# Calculate Random Forest F1 Score
rf_f1 = f1_score(y_test, rf_predictions)

print("Random Forest F1 Score :", round(rf_f1,4))
```

```python
# Display Confusion Matrix
rf_cm = confusion_matrix(y_test, rf_predictions)

print(rf_cm)
```

```python
# Plot Confusion Matrix
plt.figure(figsize=(6,5))

sns.heatmap(
    rf_cm,
    annot=True,
    fmt="d",
    cmap="Greens"
)

plt.title("Confusion Matrix - Random Forest")

plt.xlabel("Predicted Label")

plt.ylabel("Actual Label")

plt.show()
```

```python
print(classification_report(y_test, rf_predictions))
```

```python
# Calculate Prediction Probabilities
rf_probabilities = random_forest.predict_proba(X_test)[:,1]
```

```python
from sklearn.metrics import (
    accuracy_score,
    confusion_matrix,
    classification_report,
    f1_score,
    roc_curve,
    roc_auc_score
)
```

```python
# Calculate ROC Curve
fpr, tpr, threshold = roc_curve(
    y_test,
    rf_probabilities
)
```

```python
# Calculate AUC Score
auc_score = roc_auc_score(
    y_test,
    rf_probabilities
)

print("AUC Score :", round(auc_score,4))
```

```python
# Plot ROC Curve
plt.figure(figsize=(7,6))

plt.plot(
    fpr,
    tpr,
    label="Random Forest (AUC = %0.3f)" % auc_score
)

plt.plot(
    [0,1],
    [0,1],
    linestyle="--"
)

plt.xlabel("False Positive Rate")

plt.ylabel("True Positive Rate")

plt.title("ROC Curve")

plt.legend()

plt.show()
```

```python
# Display Feature Importance
importance = pd.DataFrame({

    "Feature": X.columns,

    "Importance": random_forest.feature_importances_

})

importance = importance.sort_values(
    by="Importance",
    ascending=False
)

importance.head(10)
```

```python
# Plot Top 10 Important Features
plt.figure(figsize=(8,6))

sns.barplot(
    data=importance.head(10),
    x="Importance",
    y="Feature"
)

plt.title("Top 10 Important Features")

plt.show()
```

```python
# Compare Logistic Regression and Random Forest
comparison = pd.DataFrame({

    "Model":[
        "Logistic Regression",
        "Random Forest"
    ],

    "Accuracy":[
        accuracy,
        rf_accuracy
    ],

    "F1 Score":[
        f1,
        rf_f1
    ]

})

comparison
```

```python
# Compare Model Accuracy
plt.figure(figsize=(6,4))

sns.barplot(
    data=comparison,
    x="Model",
    y="Accuracy"
)

plt.title("Accuracy Comparison")

plt.show()
```

```python
# Compare F1 Score
plt.figure(figsize=(6,4))

sns.barplot(
    data=comparison,
    x="Model",
    y="F1 Score"
)

plt.title("F1 Score Comparison")

plt.show()
```
